```python
import decisionTree as dt
from decisionTree import DecisionTree,  discretize_all  

class CustomUnpickler(pickle.Unpickler):
    def find_class(self, module, name):
        if module == "__main__":
            module = "decisionTree"
        return super().find_class(module, name)

with open('precision_best.pkl', 'rb') as f:
    tree = CustomUnpickler(f).load()

x_df_discretized = discretize_all(x_df)
X_test = x_df_discretized.values 

predictions = tree.predict(X_test)
```
